```python
# load the dataset
file_path = './Concrete_Data.xls'
data = pd.read_excel(file_path)

print("First five rows of the dataset:")
print(data.head())

# display basic information
print("\nDataset basic information:")
print(data.info())

print("\nDescriptive statistics:")
print(data.describe())

# check for missing values
print("\nMissing values check:")
print(data.isnull().sum())

# display the column names of the dataset
print("\nFeature column names:")
print(data.columns)

# check the distribution of the target variable (last column)
target_column = data.columns[-1]
plt.figure(figsize=(8, 6))
sns.histplot(data[target_column], kde=True, bins=30, color='blue')
plt.title(f'Distribution of Target Variable: {target_column}')
plt.xlabel(target_column)
plt.ylabel('Frequency')
plt.show()

# plot the correlation heatmap between features
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Dataset Description:
The **Concrete_Data** dataset is designed for studying and predicting the compressive strength of concrete based on its material composition and curing age. It is a regression dataset where the target variable represents the compressive strength in megapascal (MPa).

---

### Dataset Dimensions:
- **Number of Samples**: The dataset contains **1030 samples**.
- **Number of Features**: The dataset has **8 features** and **1 target variable**.
  - **Feature List**:
    1. **Cement (kg in a m mixture)**: Amount of cement.
    2. **Blast Furnace Slag (kg in a m mixture)**: Amount of blast furnace slag.
    3. **Fly Ash (kg in a m mixture)**: Amount of fly ash.
    4. **Water (kg in a m mixture)**: Amount of water.
    5. **Superplasticizer (kg in a m mixture)**: Amount of superplasticizer.
    6. **Coarse Aggregate (kg in a m mixture)**: Amount of coarse aggregate.
    7. **Fine Aggregate (kg in a m mixture)**: Amount of fine aggregate.
    8. **Age (days)**: The curing age of the concrete in days.
  - **Target Variable**:
    - **Concrete Compressive Strength (MPa)**: The compressive strength of the concrete.

---

### Dataset Characteristics:
1. **Correlations**: Certain features, such as cement amount, show strong correlations with the target variable (compressive strength), while others, like blast furnace slag, have weaker correlations.
2. **Target Distribution**: The target variable (compressive strength) ranges from approximately **0 MPa to 80 MPa**, with a slightly skewed distribution.
3. **No Missing Values**: The dataset is complete, with no missing values in any samples.
